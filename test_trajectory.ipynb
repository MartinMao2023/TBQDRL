```python
from trajectory_encoder import MaternEncoder, IntegrateMatern
import numpy as np
import jax.numpy as jnp
import jax
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
encoder = MaternEncoder(l=4, dt=0.05, horizon=5)
key = jax.random.PRNGKey(18222)
key, subkey = jax.random.split(key)
```

```python
# t = jnp.zeros((1,))
obs = jnp.ones((30)) 
task_state = encoder.reset(obs, subkey)

task_state = task_state.replace(deviation=jnp.zeros((2,)),
                                # task=jnp.ones((2, 10))
                                )

print(task_state.task[0])
y_values = []

for i in range(100):

    task_state = encoder.step(task_state, jnp.zeros((2,)))
    y_values.append(-task_state.deviation[0])


```

```python
print(encoder.prior_mu)
```

```python
plt.plot(jnp.arange(100) * 0.05, [k for k in y_values])
# plt.scatter(jnp.arange(1, 10) * 0.5, jnp.ones(9))
plt.show()
```
