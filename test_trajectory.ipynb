```python
from trajectory_encoder import FiniteMaternWrapper, State_info
import numpy as np
import jax.numpy as jnp
import jax
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
encoder = FiniteMaternWrapper(l=1, v_noise_var=0.0001, var=1.0)
key = jax.random.PRNGKey(1855)
key, subkey = jax.random.split(key)

state_info = State_info(current_position=jnp.zeros(2), current_velocity=jnp.ones(2))
initial_z_state = encoder._init_task_state(state_info, subkey)
```

```python
z_state = initial_z_state
y_values = []
t_values = []

for i in range(80):
    z_state, ys, ts = encoder.step(z_state) # y of (2,)
    y_values.append(ys[0])
    t_values.append(ts)
```

```python
plt.plot(jnp.arange(80) * 0.05, [k for k in y_values])
# plt.scatter(jnp.arange(1, 10) * 0.5, jnp.ones(9))
plt.show()
```
